```python
import os
import pandas as pd
import numpy as np
```

```python
func = lambda x : x.fillna(round(x.mean(), 2))

def interpolation(df):
    df = df[df["Year"] != 0]
    df.reset_index(inplace=True)
    df = df.drop(columns=["index"])
    df["Value"] = df["Value"].replace(-999.000, None)
    if not df.empty:
        df2 = df.groupby('Month')
        df2["Value"].transform(func)
    return df    
```

```python
dataframes_to_concat = []

for filename in os.listdir("aus_data2/"):
    if "Day" in filename:
        data_date_day = []
        data_date_month = []
        data_date_year = []
        data_value = []
        data_date_common = []
        grdc_no = ""
                
        f = open("aus_data2/" + filename, "rb")
        f2 = pd.read_csv("NCDC_3/" + filename[:7] + ".csv")
        
        for i in f:
            i = str(i)
            i = i[2:-5]
            if "GRDC-No" in i:
                grdc_no = i.split(":")[1].strip()
                continue
            elif "#" in i:
                continue
            temp = i.split(";")
            date_temp = temp[0].split("-")
            date_temp[0] = date_temp[0].strip()
            if date_temp[0] == "YYYY":
                date_temp[0] = "0000"
                date_temp[1] = "00"
                date_temp[2] = "00"
            data_date_year.append(int(date_temp[0].strip()))
            data_date_month.append(int(date_temp[1].strip()))
            data_date_day.append(int(date_temp[2].strip()))
            data_date_common.append(date_temp[0].strip() + "-" + date_temp[1].strip() + "-" + date_temp[2].strip())
            temp[2] = temp[2].strip()
            if temp[2] == "Value":
                temp[2] = "-999.000"
            data_value.append(round(float(temp[2].strip()), 3))
        df_temp = pd.DataFrame(list(zip(data_date_day, data_date_month, data_date_year, data_value, data_date_common)), columns =['Day', 'Month', 'Year', "Value", "Date"])
        df_temp["GRDC_No"] = int(grdc_no)
        df_temp = df_temp.set_index('Date').join(f2.set_index('DATE'))
        #df_temp = interpolation(df_temp)
        dataframes_to_concat.append(df_temp)
```

```python
df = pd.concat(dataframes_to_concat)
```
